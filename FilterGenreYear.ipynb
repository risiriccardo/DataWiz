# filtered: movie-only, startYear >= 2000, droped null in genres & runtimeMinutes

```python
import pandas as pd
```

```python
# Load the .tsv.gz file directly
df4 = pd.read_csv("title.basics.tsv.gz", compression='gzip', sep='\t')
```

```python
df4.tail(50)
```

```python
# Convert 'startYear' to numeric, setting errors='coerce' to handle non-numeric values as NaN
df4['startYear'] = pd.to_numeric(df4['startYear'], errors='coerce')

# Drop rows where 'startYear' is NaN to keep only integer values
df4 = df4.dropna(subset=['startYear'])

# Convert 'startYear' to integer after dropping NaN values
df4['startYear'] = df4['startYear'].astype(int)

# Step 1: Filter to include only rows where titleType is 'movie'
movies_df = df4[df4['titleType'] == 'movie']

# Step 2: Further filter movies_df to include only rows where startYear is greater than or equal to 2000
movies_2000_df = movies_df[movies_df['startYear'] >= 2000]

movies_2000_df.head(50)
```

```python
# Replace '\N' with NaN in both 'genres' and 'runtimeMinutes' columns
movies_2000_df['genres'].replace('\\N', pd.NA, inplace=True)
movies_2000_df['runtimeMinutes'].replace('\\N', pd.NA, inplace=True)

# Drop rows with NaN values in either 'genres' or 'runtimeMinutes' columns
movies_2000_df = movies_2000_df.dropna(subset=['genres', 'runtimeMinutes'])
```
