# filtered: movie-only, droped null in genres & runtimeMinutes

```python
import pandas as pd
```

```python
# Load the .tsv.gz file directly
df = pd.read_csv("title.basics.tsv.gz", compression='gzip', sep='\t', low_memory=False)
```

```python
df.head(50)
```

```python
#Filter to include only rows where titleType is 'movie'
movies_df = df[df['titleType'] == 'movie']

movies_df.head(50)
```

```python
# Replace '\N' with NaN in both 'genres' and 'runtimeMinutes' columns
movies_df['genres'].replace('\\N', pd.NA, inplace=True)
movies_df['runtimeMinutes'].replace('\\N', pd.NA, inplace=True)

# Drop rows with NaN values in either 'genres' or 'runtimeMinutes' columns
movies_df = movies_df.dropna(subset=['genres', 'runtimeMinutes'])
```

```python
# Display the cleaned DataFrame
movies_df.head(50)
```

```python
# Split the 'genres' column by comma and extract unique genres
unique_genres = set(genre for genres in movies_df['genres'].dropna().str.split(',') for genre in genres)

# Convert the unique genres set into a DataFrame for table display
unique_genres_df = pd.DataFrame(list(unique_genres), columns=['Unique Genres'])

# Display the table
print(unique_genres_df)
```
